```
import numpy as np
import SimpleITK as sitk
import matplotlib.pyplot as plt
from sklearn import preprocessing

#定義讀取資料的函式
def read_data(num):
  trainmask = sitk.ReadImage(path+"/xVertSeg.v1/Data1/resize_masks_128/reize_mask"+str(num)+".mhd")
  trainimage = sitk.ReadImage(path+"/xVertSeg.v1/Data1/resize_images_128/reize_image"+str(num)+".mhd")
  trainmasks=sitk.GetArrayFromImage(trainmask).astype('float32')
  trainimages=sitk.GetArrayFromImage(trainimage).astype('float32')
  print("trainmasks shape:",trainmasks.shape)
  return trainimages,trainmasks
```

```
#使用迴圈逐筆進資料
#以由上往下(Z軸)搜尋的方式切割每節椎骨
#將每個椎骨個別存檔 ex.L1、L2、L3、L4、L5

for num in range(1,16):
  trainimages,trainmasks=read_data(num)

  #L1
  bone_mask=np.zeros((128,128,64),dtype=np.float64)
  bone_image=np.zeros((128,128,64),dtype=np.float64)
  index=0
  for z in range(256):
    bone=trainmasks[:,:,z]
    if 200 in np.unique(bone):
      print("num:",num," index:",z)
      bone_mask=trainmasks[:,:,z:z+64]
      bone_mask=bone_mask.reshape(-1)
      bone_image=trainimages[:,:,z:z+64]
      bone_image=bone_image.reshape(-1)
      for i in range(len(bone_mask)):
        if bone_mask[i] !=200:
          bone_image[i]=0
      bone_image=bone_image.reshape(128,128,64)
      sitk_masks=sitk.GetImageFromArray(bone_image)
      mask_filename=path+"/xVertSeg.v1/Data1/bone_masks/bone_mask_L1_"+str(num)+".mhd"
      sitk.WriteImage(sitk_masks,mask_filename)
      index=z
      break;

  #L0
  bone_mask=np.zeros((128,128,64),dtype=np.float64)
  bone_image=np.zeros((128,128,64),dtype=np.float64)
  index=0
  for z in range(256):
    bone=trainmasks[:,:,z]
    if 210 in np.unique(bone):
      print("num:",num," index:",z)
      bone_mask=trainmasks[:,:,z:z+64]
      bone_mask=bone_mask.reshape(-1)
      bone_image=trainimages[:,:,z:z+64]
      bone_image=bone_image.reshape(-1)
      for i in range(len(bone_mask)):
        if bone_mask[i] !=210:
          bone_image[i]=0
      bone_image=bone_image.reshape(128,128,64)
      sitk_masks=sitk.GetImageFromArray(bone_image)
      mask_filename=path+"/xVertSeg.v1/Data1/bone_masks/bone_mask_L2_"+str(num)+".mhd"
      sitk.WriteImage(sitk_masks,mask_filename)
      index=z
      break;

  #L3
  bone_mask=np.zeros((128,128,64),dtype=np.float64)
  bone_image=np.zeros((128,128,64),dtype=np.float64)
  index=0
  for z in range(256):
    bone=trainmasks[:,:,z]
    if 220 in np.unique(bone):
      print("num:",num," index:",z)
      bone_mask=trainmasks[:,:,z:z+64]
      bone_mask=bone_mask.reshape(-1)
      bone_image=trainimages[:,:,z:z+64]
      bone_image=bone_image.reshape(-1)
      for i in range(len(bone_mask)):
        if bone_mask[i] !=220:
          bone_image[i]=0
      bone_image=bone_image.reshape(128,128,64)
      sitk_masks=sitk.GetImageFromArray(bone_image)
      mask_filename=path+"/xVertSeg.v1/Data1/bone_masks/bone_mask_L3_"+str(num)+".mhd"
      sitk.WriteImage(sitk_masks,mask_filename)
      index=z
      break;

  #L4
  bone_mask=np.zeros((128,128,64),dtype=np.float64)
  bone_image=np.zeros((128,128,64),dtype=np.float64)
  index=0
  for z in range(256):
    bone=trainmasks[:,:,z]
    if 230 in np.unique(bone):
      print("num:",num," index:",z)
      if z+64<256:
        bone_mask=trainmasks[:,:,z:z+64]
        bone_mask=bone_mask.reshape(-1)
        bone_image=trainimages[:,:,z:z+64]
        bone_image=bone_image.reshape(-1)
      else:
        bone_mask=trainmasks[:,:,-64:]
        bone_mask=bone_mask.reshape(-1)
        bone_image=trainimages[:,:,-64:]
        bone_image=bone_image.reshape(-1)
      for i in range(len(bone_mask)):
        if bone_mask[i] !=230:
          bone_image[i]=0
      bone_image=bone_image.reshape(128,128,64)
      sitk_masks=sitk.GetImageFromArray(bone_image)
      mask_filename=path+"/xVertSeg.v1/Data1/bone_masks/bone_mask_L4_"+str(num)+".mhd"
      sitk.WriteImage(sitk_masks,mask_filename)
      index=z
      break;

  #L5
  bone_mask=np.zeros((128,128,64),dtype=np.float64)
  bone_image=np.zeros((128,128,64),dtype=np.float64)
  index=0
  for z in range(256):
    bone=trainmasks[:,:,z]
    if 240 in np.unique(bone):
      print("num:",num," index:",z)
      if z+64<256:
        bone_mask=trainmasks[:,:,z:z+64]
        bone_mask=bone_mask.reshape(-1)
        bone_image=trainimages[:,:,z:z+64]
        bone_image=bone_image.reshape(-1)
      else:
        bone_mask=trainmasks[:,:,-64:]
        bone_mask=bone_mask.reshape(-1)
        bone_image=trainimages[:,:,-64:]
        bone_image=bone_image.reshape(-1)
      for i in range(len(bone_mask)):
        if bone_mask[i] !=240:
          bone_image[i]=0
      bone_image=bone_image.reshape(128,128,64)
      sitk_masks=sitk.GetImageFromArray(bone_image)
      mask_filename=path+"/xVertSeg.v1/Data1/bone_masks/bone_mask_L5_"+str(num)+".mhd"
      sitk.WriteImage(sitk_masks,mask_filename)
      index=z
      break;
```
