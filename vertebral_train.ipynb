```
from keras.models import load_model
from keras.optimizers import *
from sklearn.model_selection import train_test_split
import numpy as np
import SimpleITK as sitk
import matplotlib.pyplot as plt
from sklearn import preprocessing
from keras.utils import np_utils

#宣告兩個list存放image和mask
trainimages_list=[]
trainmasks_list=[]

#以迴圈逐筆讀進資料
#將image的數值範圍設定為0~255
#將label值設定成0~5
for i in range(1,16):
  print("第",i,"筆")
  trainimage = sitk.ReadImage(path+"/xVertSeg.v1/Data1/resize_images_128/reize_image"+str(i)+".mhd")
  trainmask = sitk.ReadImage(path+"/xVertSeg.v1/Data1/resize_masks_128/reize_mask"+str(i)+".mhd")

  trainimages=sitk.GetArrayFromImage(trainimage).astype('float32')
  trainimages=trainimages+(-trainimages.min())
  print('trainimages max:',trainimages.max())
  trainimages=trainimages/trainimages.max()*255
  print('trainimages max:',trainimages.max())
  trainmasks=sitk.GetArrayFromImage(trainmask)
  trainmasks_copy=trainmasks.copy()
  trainmasks_copy[trainmasks_copy==200]=1
  trainmasks_copy[trainmasks_copy==210]=2
  trainmasks_copy[trainmasks_copy==220]=3
  trainmasks_copy[trainmasks_copy==230]=4
  trainmasks_copy[trainmasks_copy==240]=5
  print('trainmasks_copy:',np.unique(trainmasks_copy))

  #增加一個通道維度
  trainimages=np.expand_dims(trainimages,axis=3)
  trainmasks=np.expand_dims(trainmasks_copy,axis=3)

  #onehot encoding
  trainmasks_onehot=np_utils.to_categorical(trainmasks,6)

  trainimages_list.append(trainimages)
  trainmasks_list.append(trainmasks_onehot)

  img_data_train=np.array(trainimages_list)
  img_data_masks=np.array(trainmasks_list)
  print('img_data_train shape:',img_data_train.shape)
  print('img_data_masks shape:',img_data_masks.shape)

model=load_model(path+'/xVertSeg.v1/model/Unetmodel.h5')

#開始訓練
#參數可自行調整
model.fit(img_data_train[0:12],img_data_masks[0:12],validation_data=(img_data_train[12:],img_data_masks[12:]),batch_size=1,epochs=500)

model.save(path+'/xVertSeg.v1/model/Unetmodel.h5')
```
